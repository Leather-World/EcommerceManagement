```python
import pandas as pd
```

```python
amazon_return_report = pd.read_csv("amazon.csv")
flipkart_return_report = pd.read_csv("flipkart.csv")
# myntra_return_report = pd.read_csv("ne/Myntra_RR_Jan1toApril30_data.csv")
ajio_return_report = pd.read_csv("ajio.csv")
```

```python
amazon_return_report = amazon_return_report[['Order ID','Return request date','Tracking ID','ASIN',' Merchant SKU','Return reason','Return type','Return delivery date']]
amazon_return_report.head(2)

amazon_return_report.rename(columns = {'Order ID':'order_id','Return request date':'return_date','ASIN':'platformID', 'Tracking ID':'tracking_id',
                                      ' Merchant SKU':'sku','Return reason':'return_reason', 'Return delivery date':'return_delivery_date'}, inplace = True)

amazon_return_report = amazon_return_report[~amazon_return_report['Return type'].isin(['Rejected', 'Undelivered'])]

amazon_return_report.drop(['Return type'], axis=1,  inplace=True)

# # convert the purchase-date column to a datetime object
amazon_return_report['return_date'] = pd.to_datetime(amazon_return_report['return_date'], format = '%d-%b-%y')

# # extract the date from the datetime object
amazon_return_report['return_date'] = amazon_return_report['return_date'].dt.strftime('%Y-%m-%d')

amazon_return_report['return_delivery_date'] = amazon_return_report['return_delivery_date'].str.strip()
# convert to datetime objects
amazon_return_report['return_delivery_date'] = pd.to_datetime(amazon_return_report['return_delivery_date'], format='%d-%b-%y')

# convert to YYYY-MM-DD format
amazon_return_report['return_delivery_date'] = amazon_return_report['return_delivery_date'].dt.strftime('%Y-%m-%d')



amazon_return_report['platform'] = 'A'
amazon_return_report['return_sub_reason'] = 0

reorder = ['order_id', 'return_date', 'tracking_id', 'platformID', 'sku', 'return_reason', 'return_sub_reason', 'return_delivery_date', 'platform']

amazon_return_report = amazon_return_report.reindex(columns=reorder)


amazon_return_report.head(2)
```

```python
amazon_return_report['return_reason'].unique()
```

```python
# flipkart_return_report = pd.read_csv("ne/flipkart.csv")

flipkart_return_report = flipkart_return_report[['order_item_id','return_approval_date','return_id','fsn','sku','return_reason','return_sub_reason','return_type', 'return_completion_date']]

flipkart_return_report.rename(columns = {'order_item_id':'order_id','return_approval_date':'return_date','fsn':'platformID', 'return_id':'tracking_id',
                                          'return_completion_date':'return_delivery_date'}, inplace = True)

flipkart_return_report = flipkart_return_report[~flipkart_return_report['return_type'].isin(['courier_return'])]

flipkart_return_report.drop(['return_type'], axis=1,  inplace=True)

# # # convert the purchase-date column to a datetime object
flipkart_return_report['return_date'] = pd.to_datetime(flipkart_return_report['return_date'])

# # # extract the date from the datetime object
flipkart_return_report['return_date'] = flipkart_return_report['return_date'].dt.strftime('%Y-%m-%d')

flipkart_return_report['return_delivery_date'] = pd.to_datetime(flipkart_return_report['return_delivery_date'])

flipkart_return_report['return_delivery_date'] = flipkart_return_report['return_delivery_date'].dt.strftime('%Y-%m-%d')

flipkart_return_report['sku'] = flipkart_return_report['sku'].str.split(':').str[1]
flipkart_return_report['order_id'] = flipkart_return_report['order_id'].str.split(':').str[1]
flipkart_return_report['tracking_id'] = flipkart_return_report['tracking_id'].str.split(':').str[1]



flipkart_return_report['platform'] = 'F'
flipkart_return_report['platformID'] = 0

reorder = ['order_id', 'return_date', 'tracking_id', 'platformID', 'sku', 'return_reason', 'return_sub_reason', 'return_delivery_date', 'platform']

flipkart_return_report = flipkart_return_report.reindex(columns=reorder)


flipkart_return_report.head(2)
```

```python
# myntra_return_report = pd.read_csv("ne/myntra.csv")

myntra_return_report = myntra_return_report[['order_id','return_created_date','return_id','style_id','return_reason','type','lmdo_last_modified_on']]
myntra_return_report.head(2)

myntra_return_report.rename(columns = {'return_created_date':'return_date','style_id':'platformID', 'return_id':'tracking_id',
                                       'lmdo_last_modified_on':'return_delivery_date'}, inplace = True)

myntra_return_report = myntra_return_report[~myntra_return_report['type'].isin(['RTO'])]

myntra_return_report.drop(['type'], axis=1,  inplace=True)

# # # # convert the purchase-date column to a datetime object
myntra_return_report['return_date'] = pd.to_datetime(myntra_return_report['return_date'], format='mixed')

# # # # extract the date from the datetime object
myntra_return_report['return_date'] = myntra_return_report['return_date'].dt.strftime('%Y-%m-%d')

# # # # convert the purchase-date column to a datetime object
myntra_return_report['return_delivery_date'] = pd.to_datetime(myntra_return_report['return_delivery_date'])

# # # # extract the date from the datetime object
myntra_return_report['return_delivery_date'] = myntra_return_report['return_delivery_date'].dt.strftime('%Y-%m-%d')

myntra_return_report['platform'] = 'M'
myntra_return_report['sku'] = 0
myntra_return_report['return_sub_reason'] = 0

myntra_return_report['tracking_id'] = myntra_return_report['tracking_id'].apply(lambda x: '{:.0f}'.format(x))


reorder = ['order_id', 'return_date', 'tracking_id', 'platformID', 'sku', 'return_reason', 'return_sub_reason', 'return_delivery_date', 'platform']

myntra_return_report = myntra_return_report.reindex(columns=reorder)


myntra_return_report.head(2)
```

```python
ajio_return_report = ajio_return_report[['Return Order No', 'Cust Return Created Date', 'EAN', 'JioCode', 'SELLER_SKU','Return Reason', 'Return type','Return Sub reason']]
ajio_return_report.head()


ajio_return_report.rename(columns = {'Return Order No':'order_id','Cust Return Created Date':'return_date','JioCode':'platformID', 'EAN':'tracking_id',
                                      'SELLER_SKU':'sku','Return Reason':'return_reason','Return Sub reason':'return_sub_reason'}, inplace = True)

ajio_return_report = ajio_return_report[~ajio_return_report['Return type'].isin(['Self Ship Without Shipping Charge'])]

ajio_return_report.drop(['Return type'], axis=1,  inplace=True)

# parse the date strings using dateutil
ajio_return_report['return_date'] = ajio_return_report['return_date'].apply(parser.parse)

# convert to YYYY-MM-DD format
ajio_return_report['return_date'] = ajio_return_report['return_date'].dt.strftime('%Y-%m-%d')

ajio_return_report['platform'] = 'AJ'
ajio_return_report['return_delivery_date'] = 0

reorder = ['order_id', 'return_date', 'tracking_id', 'platformID', 'sku', 'return_reason', 'return_sub_reason', 'return_delivery_date', 'platform']

ajio_return_report = ajio_return_report.reindex(columns=reorder)


ajio_return_report.head(2)
```
